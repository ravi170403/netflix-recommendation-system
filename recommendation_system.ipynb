```python
import pandas as pd

ratings = pd.read_csv(
    "Data/u.data",
    sep="\t",
    names=["user_id","movie_id","rating","timestamp"]
)

ratings.head()
```

```python
movies = pd.read_csv(
    "Data/u.item",
    sep="|",
    encoding="latin-1",
    names=[
        "movie_id","title","release_date","video_release_date",
        "IMDb_URL","unknown","Action","Adventure","Animation",
        "Children","Comedy","Crime","Documentary","Drama",
        "Fantasy","Film-Noir","Horror","Musical","Mystery",
        "Romance","Sci-Fi","Thriller","War","Western"
    ]
)

movies.head()
```

```python
data = pd.merge(ratings, movies, on="movie_id")

data.head()
```

```python
movie_ratings = data.pivot_table(
    index="user_id",
    columns="title",
    values="rating"
)

movie_ratings.head()
```

```python
movie_ratings.shape
```

```python
movie_ratings.isnull().sum().head()
```

```python
toy_story_ratings = movie_ratings["Toy Story (1995)"]

toy_story_ratings.head()
```

```python
similar_movies = movie_ratings.corrwith(toy_story_ratings)
```

```python
corr_toy_story = pd.DataFrame(similar_movies, columns=["Correlation"])
corr_toy_story.dropna(inplace=True)
corr_toy_story.head()
```

```python
ratings_count = data.groupby("title")["rating"].count()
corr_toy_story["num_ratings"] = ratings_count
corr_toy_story.head()
```

```python
recommendations = corr_toy_story[corr_toy_story["num_ratings"] > 100] \
                    .sort_values("Correlation", ascending=False)

recommendations.head(10)
```

```python
def recommend_movies(movie_name):

   

    movie_user_ratings = movie_ratings[movie_name]

    similar_movies = movie_ratings.corrwith(movie_user_ratings)

    corr_df = pd.DataFrame(similar_movies, columns=["Correlation"])
    corr_df.dropna(inplace=True)

    ratings_count = data.groupby("title")["rating"].count()
    corr_df["num_ratings"] = ratings_count

    recommendations = corr_df[corr_df["num_ratings"] > 100] \
                        .sort_values("Correlation", ascending=False)

    return recommendations.head(10)
```

```python
recommend_movies("Toy Story (1995)")
```

```python
def recommend_movies(movie_name):

    movie_user_ratings = movie_ratings[movie_name]

    similar_movies = movie_ratings.corrwith(movie_user_ratings)

    corr_df = pd.DataFrame(similar_movies, columns=["Correlation"])
    corr_df.dropna(inplace=True)

    ratings_count = data.groupby("title")["rating"].count()
    corr_df["num_ratings"] = ratings_count

    recommendations = corr_df[corr_df["num_ratings"] > 100] \
                        .sort_values("Correlation", ascending=False)

    # ✅ REMOVE SAME MOVIE
    recommendations = recommendations.drop(movie_name)

    return recommendations.head(10)
```

```python
recommend_movies("Toy Story (1995)")
```
